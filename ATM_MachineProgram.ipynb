```python
# Program an ATM Machine Mechanism flow

# Prompt the user to insert their card
print("Insert Your Card")

# Prompt the user to enter their 4-digit PIN and convert the input to an integer
EnterPin = int(input("Enter Your 4 Digit Pin: "))

# Display the menu options and prompt the user to make a selection
menu = input('''
MENU
1. Check Balance
2. Deposit Funds
3. Withdraw Funds
4. Exit
''')

# Validate the user's menu selection; continue prompting until a valid selection is made
while menu not in ['1', '2', '3', '4']:
    # Inform the user of an invalid selection and prompt again
    print(f'Invalid Selection please start again:')
    menu = input('''
    1. Check Balance
    2. Deposit Funds
    3. Withdraw Funds
    4. Exit
    ''')

# Perform actions based on the user's menu selection
if menu == '1':
    # User selected "Check Balance"
    menu_type = 'Check Balance'
    amount = EnterPin  # Assuming EnterPin represents the account balance
    # Display the current balance formatted with commas and two decimal places
    print(f'Your current balance is: PKR.{EnterPin:,.2f}')

elif menu == '2':
    # User selected "Deposit Funds"
    menu_type = 'Deposit Funds'
    # Prompt the user to enter the deposit amount
    print(f'Please enter your deposit amount:')
    amount = int(input())
    # Confirm the deposited amount to the user with proper formatting
    print(f'Thank you for depositing "PKR.{amount:,.2f}"')
    # Simulate printing a receipt
    print(f'Printing your receipt........')
    # Thank the user and prompt them to take their card
    print(f'Please take your card. Thank you for banking with us!')

elif menu == '3':
    # User selected "Withdraw Funds"
    menu_type = 'Withdraw Funds'
    # Prompt the user to enter the withdrawal amount
    amount = int(input("Enter the amount: "))
    # Simulate processing the withdrawal
    print(f'Please wait........')
    # Inform the user that the cash is ready to be taken with proper formatting
    print(f'Please take your cash "PKR.{amount:,.2f}"')
    # Simulate printing a receipt
    print(f'Printing your receipt......')
    # Thank the user and prompt them to take their card
    print(f'Please take your card. Thank you for banking with us!')

elif menu == '4':
    # User selected "Exit"
    menu_type = 'Exit'
    # Thank the user and prompt them to take their card
    print(f'{menu_type} - Take your card. Thank you for banking with us.')

else:
    # Handle any unexpected cases (although this should not occur due to prior validation)
    print('Please take your card.')
```
