```python
# Display coffee options and take input for coffee type
CoffeeType = input('''
1. Espresso
2. Americano
3. Latte
4. Cappuccino
''')

# Validate the coffee type input; if not in the options, ask again
while CoffeeType not in ['1', '2', '3', '4']:
  # Prompt the user to choose the correct option if the input is invalid
  print(f'Please Choose Correct Selection: {CoffeeType}')
  # Re-display the coffee options and take input again
  CoffeeType = input('''
1. Espresso
2. Americano
3. Latte
4. Cappuccino
''')

# Assign the appropriate coffee name based on the user's selection
if CoffeeType == '1':
  CoffeeName = 'Espresso'
elif CoffeeType == '2':
  CoffeeName = 'Americano'
elif CoffeeType == '3':
  CoffeeName = 'Latte'
elif CoffeeType == '4':
  CoffeeName = 'Cappuccino'

# Display the selected coffee name
print(f'Coffee Name: {CoffeeName}')

# Ask the user to choose the size of the coffee
CoffeeSize = input('''
Choose the size of your coffee:
1. Small
2. Medium
3. Large
''')

# Validate the coffee size input; if not in the options, ask again
while CoffeeSize not in ['1', '2', '3']:
  # Re-prompt for coffee size if the input is invalid
  CoffeeSize = input('''
Choose the size of your coffee:
1. Small
2. Medium
3. Large
''')

# Assign the appropriate size based on the user's selection
if CoffeeSize == '1':
  Size = 'Small'
elif CoffeeSize == '2':
  Size = 'Medium'
elif CoffeeSize == '3':
  Size = 'Large'

# Display the selected coffee name along with the size
print(f'{CoffeeName} {Size}')

# Ask the user if they want to add sugar
SugarAdd = input('''
Do you want to add sugar:
1. Yes
2. No
''')

# Assign the sugar preference based on the user's selection
if SugarAdd == '1':
  sugar = 'Yes'
elif SugarAdd == '2':
  sugar = 'No'

# Display the sugar preference
print(sugar)

# Ask the user if they want to add milk
MilkAdd = input('''
Do you want to add milk:
1. Yes
2. No
''')

# Assign the milk preference based on the user's selection
if MilkAdd == '1':
  milk = 'Yes'
elif MilkAdd == '2':
  milk = 'No'

# Display the milk preference
  print(milk)

# If the milk input is invalid, prompt to start again
else:
  print(f'Invalid input please start again.')

# Final message indicating the coffee is being prepared
print(f'Preparing your {Size} {CoffeeName} .......')

# Final message indicating the coffee is ready
print(f'Your {Size} {CoffeeName} is ready! Enjoy your coffee!')
```
